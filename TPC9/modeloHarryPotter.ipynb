```python
from gensim.models import Word2Vec
import spacy
nlp = spacy.load("pt_core_news_sm")
```

```python
f_1= open("HPLivros/Harry Potter e A Pedra Filosofal.txt", "r", encoding='utf-8')
primeiro_livro= f_1.read()
livro1= nlp(primeiro_livro)
```

```python
f_2= open("HPLivros/Harry_Potter_Camara_Secreta-br.txt", "r", encoding='utf-8')
segundo_livro= f_2.read()
livro2= nlp(segundo_livro)
```

```python
frases = []

for doc in [livro1, livro2]:
    for sent in doc.sents:
        palavras = []
        for token in sent:
            if not token.is_punct and not token.is_space:
                palavras.append(token.text.lower())

        if len(palavras) > 0:
            frases.append(palavras)
```

```python
model=Word2Vec(frases, vector_size=100, window=5, min_count=3, sg=0, epochs=20, workers=3)
```
