1. given a string “s”, reverse it.


```python
def inverso(s):
    return s[::-1]
```

```python
inverso("arvore")
```

2. given a string “s”, returns how many “a” and “A” characters are present in it.


```python
def numDeAs(s):
    res=0
    for letra in s:
        if letra == "A" or letra == "a":
            res +=1
    return res
```

```python
numDeAs("Autocarro")
```

3. given a string “s”, returns the number of vowels there are present in it.


```python
def numVogais(s):
    vogais="aeiou"
    res=0
    minuscula= s.lower()
    for letra in minuscula:
        if letra in vogais:
            res +=1
    return res
```

```python
numVogais("ornitorrinco")
```

6. Verifica se uma string é capicua


```python
def capicua(s):
    if s == s[::-1]:
        return True 
    return False
```

```python
capicua("ana")
```

7. Verifica se duas strings estão balanceadas (Duas strings, s1 e s2, estão
balanceadas se todos os caracteres de s1 estão presentes em s2.)


```python
def balanceadas(s1,s2):
    for elem in s1:
        if elem not in s2:
            return False
    return True
```

```python
balanceadas("ana", "anabel")
```

8. Calcula o número de ocorrências de s1 em s2

```python
def numOcorr(s1,s2):
    res=0
    for i in range(len(s2)):
        if s2[i:i+len(s1)] == s1:
            res+=1
    return res
```

```python
numOcorr("tia", "matiastia")
```

9. Verifica se s1 é anagrama de s2.
○ "listen" e "silent": Deve imprimir True
○ "hello", "world": Deve imprimir False

```python
def anagrama(s1, s2):
    if len(s1) != len(s2):
        return False
    for letra in s1:
        if letra not in s2:
            return False
        else:
            s2=s2.replace(letra, "", 1) #elimina a primeira instancia de "letra" que existir em s2
    return True
```
